- Haciendo el análisis exploratorio nos deberíamos haber dado cuenta de que algunas de las variables (danceability, energy, key, loudness, speechiness, acousticness, instrumentalness) no son del tipo que deberían. Esto es debido a que los decimales están establecidos como comas y no con puntos. Cread una función que nos permita cambiar esas comas por puntos para que los datos tengan el tipo correcto.

```python
def int_a_float(c):
    '''
    Funcion que remplaza las comas por puntos de los datos del dataframe y los convierte a numeros decimales

    Parámetro: cada dato de la columna dada
    Return: el dato convertido a numero decimal
    '''
    try:
        return float(c.replace(',','.'))
    except:
        np.nan
```

```python
columnas_float = ['danceability', 'energy', 'key', 'loudness', 'speechiness', 'acousticness', 'instrumentalness']
```

```python
for col in df[columnas_float ]:
   df[col] = df[col].apply(int_a_float)
```

```python
df[columnas_float].sample()
```

- Hay algunas canciones cuyo título está en mayúsulas. Crea una función para que todos los títulos estén en minúscula. Haced lo mismo para las columnas de "track" y "album".

```python
df['title'].unique()
```

```python
def titulos(t):
    '''
    Funcion que unifica el formato de los títulos, poniendo la primera letra de cada palabra 
    en mayúsculas y el resto en minúscula

    Parámetro: cada dato de la columna dada (titulos)
    Return: el titulo en el formato deseado
    '''
    try:
        return t.title()
    except:
        return np.nan
```

```python
columnas_titulos = ['title', 'track', 'album']
```

```python
for col in df[columnas_titulos]:
   df[col] = df[col].apply(titulos)
```

```python
df[columnas_titulos].sample(3)
```

- Algunos de los artistas tienen símbolos raros, en concreto el símbolo $. Cread una función que nos elimine este símbolo de los nombres de los artistas.

```python
# analizamos en qué nombres de artista se encuentra el símbolo erroneo
n_list = []
for n in df['artist']:
    if '$' in n:
        n_list.append(n)

df_nlist = pd.DataFrame(n_list)
df_nlist[0].unique()
```

```python
# Vemos que en algunos nombres de artista hay un $ en vez de una S (ej. 'Too $hort'), por lo que creamos una funcion que lleve a cabo dicho cambio
def nombre_artista(a):
    '''
    Funcion que elimina el símbolo $ de los nombres de los artistas

    Parámetro: cada dato de la columna dada (artistas)
    Return: el nombre del artista en el formato deseado   
    '''
    try:
        return a.replace('$','S').title()
    except:
        return np.nan
```

```python
df['artist'] = df['artist'].apply(nombre_artista)
```

```python
# Comprobamos que los nombres se han cambiado correctamente
df[df['artist'] == 'Currensy'].sample()
```

```python
album_coef = df_album['views'].std()/df_album['views'].mean()
single_coef = df_single['views'].std()/df_single['views'].mean()
compilation_coef = df_compilation['views'].std()/df_compilation['views'].mean()
```

Comprobamos que el coeficiente de dispersión es elevado, confirmando nuestra intuicón de que los datos en los tres grupos son muy homogeneos.
