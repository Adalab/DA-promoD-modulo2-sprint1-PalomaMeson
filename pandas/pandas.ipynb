- Haciendo el análisis exploratorio nos deberíamos haber dado cuenta de que algunas de las variables (danceability, energy, key, loudness, speechiness, acousticness, instrumentalness) no son del tipo que deberían. Esto es debido a que los decimales están establecidos como comas y no con puntos. Cread una función que nos permita cambiar esas comas por puntos para que los datos tengan el tipo correcto.

```python
def int_a_float(c):
    '''
    Funcion que remplaza las comas por puntos de los datos del dataframe y los convierte a numeros decimales

    Parámetro: cada dato de la columna dada
    Return: el dato convertido a numero decimal
    '''
    try:
        return float(c.replace(',','.'))
    except:
        np.nan
```

```python
# creamos una lista con los nombres de las columnas para hacer el cambio de manera más eficiente
columnas_float = ['danceability', 'energy', 'key', 'loudness', 'speechiness', 'acousticness', 'instrumentalness']
```

```python
for col in df[columnas_float ]:
   df[col] = df[col].apply(int_a_float)
```

- Después de haber hecho los cambios, chequead los tipos de datos. ¿Son ya del tipo correcto? En caso de que no, cread otra función o modificar la anterior para que sean de tipo float.

```python
# y comprobamos que el cambio se ha realizado correctamente
df[columnas_float].sample(3)
```

- Hay algunas canciones cuyo título está en mayúsulas. Crea una función para que todos los títulos estén en minúscula. Haced lo mismo para las columnas de "track" y "album".

```python
df['title'].unique()
```

```python
def titulos(t):
    '''
    Funcion que unifica el formato de los datos, poniendo la primera letra de cada palabra 
    en mayúsculas y el resto en minúscula

    Parámetro: cada dato de la columna dada
    Return: el titulo en el formato deseado
    '''
    try:
        return t.title()
    except:
        return np.nan
```

```python
# creamos una lista con el nombre de las columnas que queremos cambiar
columnas_titulos = ['title', 'track', 'album']
```

```python
# utilizamos el apply en un bucle for:
for col in df[columnas_titulos]:
   df[col] = df[col].apply(titulos)
```

```python
# comprobamos que los cambios se han realizado correctamente
df[columnas_titulos].sample(3)
```

- Algunos de los artistas tienen símbolos raros, en concreto el símbolo $. Cread una función que nos elimine este símbolo de los nombres de los artistas.

```python
# analizamos en qué nombres de artista se encuentra el símbolo erroneo
n_list = []
for n in df['artist']:
    if '$' in n:
        n_list.append(n)

df_nlist = pd.DataFrame(n_list)
df_nlist[0].unique()
```

```python
# Vemos que en algunos nombres de artista hay un $ en vez de una S (ej. 'Too $hort'), por lo que creamos una funcion que lleve a cabo dicho cambio
def nombre_artista(a):
    '''
    Funcion que elimina el símbolo $ de los nombres de los artistas y lo intercambia por una 'S'

    Parámetro: cada dato de la columna dada (artistas)
    Return: el nombre del artista en el formato deseado   
    '''
    try:
        return a.replace('$','s').title()
    except:
        return np.nan
```

```python
df['artist'] = df['artist'].apply(nombre_artista)
```

```python
# Comprobamos que los nombres se han cambiado correctamente
df[df['artist'] == 'Currensy'].sample()
```

Utilizando el dataframe original, queremos explorar si el tipo de álbum influye en la cantidad de vistas que recibe en una plataforma de transmisión de música. Para ello, debemos generar un boxplot con las variables necesarias y responder las siguientes preguntas:

```python
# colocamos las tres en un solo boxplot
plt.figure(figsize = (20,9))

fig = sns.boxplot(x = 'album_type', y = 'views', data = df)

plt.show();
```

- ¿Qué tipo de álbumes tienden a tener más vistas en promedio?

Observando la gráfica de boxplot a primera vista podría decirse que los tres tipos de albumes tienen un promedio de visitas similar, siendo en los outliers o valores extremos donde los albumes de tipo 'album' se desmarcan con numeros de visitas muy por encimna del resto.

No obstante, vamos a calcular el número de visitas totales para cada tipo de album y posteriormente el porcentaje que ese numero de visitas es respecto al total de las mismas:

```python
# calculamos el total de visitas acumuladas por tipo de album
df_type_views = df.groupby('album_type')['views'].sum().reset_index()
df_type_views
```

```python
# convertimos el numero de visitas a porcentaje
df_type_views['porcentaje'] = df_type_views['views']/df_type_views['views'].sum()
```

```python
df_type_views
```

Tras convertir el numero de visitas a porcentaje comrpobamos que el tipo de album **sí** influye en el número de visitas, siendo los de tipo **album** los que más visitas reciben (más del 75%)

- ¿Hay algún tipo de álbumes que tenga una mayor dispersión en las vistas recibidas 

Con la gráfica de boxplot podemos intuir que para los tres tipos de album, los datos son muy heterogéneos, encontrándose la mayor parte de los datos en numeros de vista más bajos pero con muchos outliers que se salen de la distribución habitual.

No obstante, vamos a calcular los estadísticos principales comprobar si estas observaciones son acertadas: 

```python
df_album['views'].agg(['mean', 'median', 'std', 'max', 'min']).reset_index().T
```

```python
df_single['views'].agg(['mean', 'median', 'std', 'max', 'min']).reset_index().T
```

```python
df_compilation['views'].agg(['mean', 'median', 'std', 'max', 'min']).reset_index().T
```

Vemos que en los tres casos existe una gran diferencia entre la media y la mediana, asi como una desviación típica elevada, lo que confirma la elevada heterogeneidad de los datos.

Calculamos el coeficiente de dispersión para terminar de confirmar nuestras sospechas:

```python
album_coef = df_album['views'].std()/df_album['views'].mean()
single_coef = df_single['views'].std()/df_single['views'].mean()
compilation_coef = df_compilation['views'].std()/df_compilation['views'].mean()
```

```python
print(album_coef)
print(single_coef)
print(compilation_coef)
```

Todos ellos son muy altos y de un tamaño similar entre ellos, lo que una vez más nos indica que la dispersión de los datos en los tres tipos de albumes es similar.
